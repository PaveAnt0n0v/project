```python
def floor_group(row):# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–º–µ–Ω—ã —ç—Ç–∞–∂–∞ –Ω–∞ –ø–µ—Ä–≤—ã–π/–ø–æ—Å–ª–µ–¥–Ω–∏–π/–¥—Ä—É–≥–æ–π
    
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        
        return '–¥—Ä—É–≥–æ–π'
```
